4. Qual o percentual de sobreviventes da primeira e terceira classes?

```python

classe1_tot = titanic[titanic['class']=='First']
classe3_tot = titanic[titanic['class']=='Third']

classe1_sob = classe1_tot[classe1_tot['alive']=='yes']
classe3_sob = classe3_tot[classe3_tot['alive']=='yes']

totclass1 = len(classe1_tot)
totclass3 = len(classe3_tot)

totclass1_sob = len(classe1_sob)
totclass3_sob = len(classe3_sob)

perc_1class = totclass1_sob / totclass1
perc_3class = totclass3_sob / totclass3

print('Percentual de sobreviventes na primeira classe {}'.format(perc_1class))
print('Percentual de sobreviventes na terceira classe {}'.format(perc_3class))

#classe3 = titanic[titanic.embarked=='first']
```

5. Analise a média de idade dos passageiros por classe e sexo (sugestão: empregue groupby). Você pode notar que?

```python
display(titanic.groupby('class').agg(['mean']))
display(titanic.groupby('sex').agg(['mean']))

analise = titanic.groupby(by=['class'])['sex'].apply(list)
display(analise.groupby('sex').agg(['mean']))
print(analise)
```
